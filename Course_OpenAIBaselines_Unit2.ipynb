## Why not pure OpenAI?

**OpenAI** is a great platform. But it's very cumbersome to use for Gazebo simulations that use ROS. There are many layers to work on and the existing examples are for different simulators or mix different development layers too much, which make it difficult to develop code.

That's where the **OpenAI ROS structure** comes in. It allows us to harness the power of OpenAI, Gazebo, and ROS, but you only code for what you need. If you are an OpenAI developer, you know nothing about Gazebo and ROS. Therefore, by using this infrastructure, you will be able to test your own AI algorithms without caring about how the simulation or ROS works. Just follow the Robot API that you want to use and you are good to go.
